# Plotting a Line Chart From Pandas Object

```python
address = "/content/mtcars.csv"
cars = pd.read_csv(address)
```

```python
cars.describe()
```

```python
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
mpg = cars['mpg']
mpg.plot()
```

```python
df = cars[['cyl', 'wt', 'mpg']]
df.plot()
```

# Time Series

```python
superset = '/content/Superstore-Sales.csv'
```

```python
sales = pd.read_csv(superset, index_col="Order Date", encoding='cp1252', parse_dates=True)
```

```python
sales.head()
```

```python
sales['Order Quantity'].plot()
```

```python
sales2 = sales.sample(n=100, random_state=25, axis=0)
plt.xlabel('Order Data')
plt.ylabel('Order Quantity')
plt.title('superstore sales')
sales2['Order Quantity'].plot()
```
