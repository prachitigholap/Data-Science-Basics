# Chi Square Test

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(cyl, am)
chi2, p, dof, expected = chi2_contingency(table.values)
print('chi-square statistics %0.3f p-value %0.3f' % (chi2, p))
```
