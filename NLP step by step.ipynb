```python
import nltk
```

```python
text = "artificial intelligence (AI), the ability of a digital computer or computer-controlled robot to perform tasks commonly associated with intelligent beings. The term is frequently applied to the project of developing systems endowed with the intellectual processes characteristic of humans, such as the ability to reason, discover meaning, generalize, or learn from past experience. Since the development of the digital computer in the 1940s, it has been demonstrated that computers can be programmed to carry out very complex tasks—such as discovering proofs for mathematical theorems or playing chess—with great proficiency. Still, despite continuing advances in computer processing speed and memory capacity, there are as yet no programs that can match full human flexibility over wider domains or in tasks requiring much everyday knowledge. On the other hand, some programs have attained the performance levels of human experts and professionals in performing certain specific tasks, so that artificial intelligence in this limited sense is found in applications as diverse as medical diagnosis, computer search engines, voice or handwriting recognition, and chatbots."
```

```python
nltk.download('punkt')
```

Sentence Tokenizer

```python
from nltk.tokenize import sent_tokenize
```

```python
sent_tk = sent_tokenize(text)
print("Sentence tokenizing the text: \n")
print(sent_tk)
```

Word Tokenizer

```python
from nltk.tokenize import word_tokenize
```

```python
word_tk = word_tokenize(text)
print("Word tokenizing the Sent: \n")
print(word_tk)
```

Removing Stop Words

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

sw = set(stopwords.words("english"))

print("These are stop words in English: \n")
print(sw)
```

```python
filtered_words = [ w for w in word_tk if not w in sw]
print("The text after removing the stop words: \n")
print(filtered_words)
```

Stemming

```python
from nltk.stem import PorterStemmer
port_stem = PorterStemmer()
```

```python
stemmed_words = []

for i in filtered_words:
  stemmed_words.append(port_stem.stem(i))

print("Filtered Sentence: \n", filtered_words, "\n")
print("Stemmed Sentence: \n", stemmed_words)
```

Lemmetizing

```python
nltk.download('wordnet')
```

```python
from nltk.stem.wordnet import WordNetLemmatizer

lem = WordNetLemmatizer()

from nltk.stem.porter import PorterStemmer
stem = PorterStemmer

lem_words = []

for i in range(len(filtered_words)):
  lem_words.append(lem.lemmatize(filtered_words[i]))

print(lem_words)
```

Parts of Speech Tagging

```python
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk import pos_tag
pos_tagged_words = pos_tag(word_tk)

print(pos_tagged_words)
```

Frequency Distribution Plots

```python
from nltk.probability import FreqDist
fd = FreqDist(word_tk)
print(fd)
```

```python
import matplotlib.pyplot as plt
fd.plot(30, cumulative=False)
plt.show()
```

```python
ld = FreqDist(lem_words)
print(fd)
ld.plot(30, cumulative=False)
plt.show()
```
